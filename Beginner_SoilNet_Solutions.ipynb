### Umbennen der Spalten
4. Die Spalten sollen entsprechend der Bodentiefe (Tiefe: -7.5/-27.5 cm) umbenannt werden:

```python
df.rename(columns={'SM1':'SM1 (Tiefe: -7.5 cm)','ST1':'ST1 (Tiefe: -7.5 cm)'}, inplace=True)
df.rename(columns={'SM2':'SM2 (Tiefe: -27.5 cm)','ST2':'ST2 (Tiefe: -27.5 cm)'}, inplace=True)
df.rename(columns={'SM3':'SM3 (Tiefe: -7.5 cm)','ST3':'ST3 (Tiefe: -7.5 cm)'}, inplace=True)
df.rename(columns={'SM4':'SM4 (Tiefe: -27.5 cm)','ST4':'ST4 (Tiefe: -27.5 cm)'}, inplace=True)
df.rename(columns={'SM5':'SM5 (Tiefe: -7.5 cm)','ST5':'ST5 (Tiefe: -7.5 cm)'}, inplace=True)
df.rename(columns={'SM6':'SM6 (Tiefe: -27.5 cm)','ST6':'ST6 (Tiefe: -27.5 cm)'}, inplace=True)
df
```

#### 5. Rückbenennung der Spalten:

```python
df.rename(columns={'SM1 (Tiefe: -7.5 cm)':'SM1','ST1 (Tiefe: -7.5 cm)':'ST1'}, inplace=True)
df.rename(columns={'SM2 (Tiefe: -27.5 cm)':'SM2','ST2 (Tiefe: -27.5 cm)':'ST2'}, inplace=True)
df.rename(columns={'SM3 (Tiefe: -7.5 cm)':'SM3','ST3 (Tiefe: -7.5 cm)':'ST3'}, inplace=True)
df.rename(columns={'SM4 (Tiefe: -27.5 cm)':'SM4','ST4 (Tiefe: -27.5 cm)':'ST4'}, inplace=True)
df.rename(columns={'SM5 (Tiefe: -7.5 cm)':'SM5','ST5 (Tiefe: -7.5 cm)':'ST5'}, inplace=True)
df.rename(columns={'SM6 (Tiefe: -27.5 cm)':'SM6','ST6 (Tiefe: -27.5 cm)':'ST6'}, inplace=True)
df
```

### Filtern
Um die doppelten Timestamps zu finden, können sich mittels _pd.set_option('display.max_rows','MAXIMALE ANZAHL DER REIHEN')_  
zunächst alle Zeilen anzeigt werden lassen.

```python
pd.set_option('display.max_rows', 250)
df
```

6. Die Timestamps: 2015-12-26 03:48:00 und 2015-12-26 12:00:00 sind doppelt.  

7. Dementsprechend kann der Filter angepasst werden.

```python
filt=((df['d_t']>'2015-12-26 03:47:59')&(df['d_t']<'2015-12-26 03:48:01')|(df['d_t']>'2015-12-26 11:59:59')&(df['d_t']<'2015-12-26 12:00:01'))
df.loc[filt]
```

Alternativ kann der Filter auch mit der Oder-Operation (|) und Identisch-Operation (==) geschrieben werden.

```python
filt=((df['d_t']=='2015-12-26 03:48:00')|(df['d_t']=='2015-12-26 12:00:00'))
df.loc[filt]
```

9. Die Filterfunktion verändert die Anzahl der Spalten und Zeilen (Data-Frame) nicht.  
Es sind nach wie vor 239 Zeilen und 13 Spalten vorhanden.

```python
df.shape
```

### Duplikate entfernen
Entferne die doppelten Timestamps: 

```python
df = df.drop_duplicates(subset='d_t', keep='first')
```

### Darstellen der Ergebnisse

10. Erstellung des Diagramms mit der Datenreihen 'SM4' und 'SM6' 

Zunächst muss das notwendige Matplotlib Package importiert werden, hierbei wird ein Kürzel (_plt_) verwendet:

```python
from matplotlib import pyplot as plt
```

#### Subplots

```python
fig1, ax1 = plt.subplots(figsize=(10,5))
```

#### Titel

```python
fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.set_title('Soil Moisture 4 and 6 (26.12.2015)', fontweight='bold')
```

#### Achsenbeschriftung und Plotten der Datenreihe

```python
fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.set_title('Soil Moisture 4 and 6 (26.12.2015)', fontweight='bold')
ax1.set_xlabel('Date [YYYY-MM-DD HH:MM:SS]')
ax1.set_ylabel('Soil Moisture [%]')
ax1.set_xticks([0,118,236])
ax1.plot(df['d_t'],df['SM4'], label='Soil Moisture 4 -27.5 cm  [%]')
```

#### Hinzufügen der zweiten Y-Achse

```python
fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.set_title('Soil Moisture 4 and 6 (26.12.2015)', fontweight='bold')
ax1.set_xlabel('Date [YYYY-MM-DD HH:MM:SS]')
ax1.set_ylabel('Soil Moisture [%]')
ax1.set_xticks([0,118,236])
ax1.plot(df['d_t'],df['SM4'], label='Soil Moisture 4 -27.5 cm  [%]')

ax2=ax1.twinx()
ax2.set_ylabel('Soil Moisture [%]')
ax2.set_xticks([0,118,236])
ax2.plot(df['d_t'],df['SM6'], color='red', label='Soil Moisture 6 -27.5 cm  [%]')
```

#### Legende

```python
fig1, ax1 = plt.subplots(figsize=(10,5))
ax1.set_title('Soil Moisture 4 and 6 (26.12.2015)', fontweight='bold')
ax1.set_xlabel('Date [YYYY-MM-DD HH:MM:SS]')
ax1.set_ylabel('Soil Moisture [%]')
ax1.set_xticks([0,118,236])
ax1.plot(df['d_t'],df['SM4'], color='black', label='Soil Moisture 4 -27.5 cm  [%]')

ax2=ax1.twinx()
ax2.set_ylabel('Soil Moisture [%]')
ax2.set_xticks([0,118,236])
ax2.plot(df['d_t'],df['SM6'], color='grey', label='Soil Moisture 6 -27.5 cm  [%]')
fig1.legend(bbox_to_anchor=(0.8,0.45,0.5,0.5))
```

#### 11. Darstellung aller Messungen vom 26.12.2015

```python
fig2, ax1 = plt.subplots(figsize=(10,5))

ax1.set_title('Soil Moisture and Soil Temperature Mix (26.12.2015)', fontweight='bold')
ax1.set_xlabel('Date [YYYY-MM-DD HH:MM:SS]')
ax1.set_ylabel('Soil Moisture [%]')
ax1.set_xticks([0,118,236])
ax1.plot(df['d_t'],df['SM1'],color='darkred', label='Soil Moisture 1 -7.5 cm [%]')
ax1.plot(df['d_t'],df['SM2'],color='darkgreen', label='Soil Moisture 2 -27.5 cm [%]')
ax1.plot(df['d_t'],df['SM3'],color='teal', label='Soil Moisture 3 -7.5 cm [%]')
ax1.plot(df['d_t'],df['SM4'],color='royalblue', label='Soil Moisture 4 -27.5 cm [%]')
ax1.plot(df['d_t'],df['SM5'],color='darkmagenta', label='Soil Moisture 5 -7.5 cm [%]')
ax1.plot(df['d_t'],df['SM6'],color='darkgrey', label='Soil Moisture 6 -27.5 cm [%]')

ax2=ax1.twinx()
ax2.set_ylabel('Soil Temperature [°C]')
ax2.set_xticks([0,118,236])
ax2.plot(df['d_t'],df['ST1'], color='red',label='Soil Temperature 1 -7.5 cm [°C]')
ax2.plot(df['d_t'],df['ST2'], color='green',label='Soil Temperature 2 -27.5 cm [°C]')
ax2.plot(df['d_t'],df['ST3'], color='cyan',label='Soil Temperature 3 -7.5 cm [°C]')
ax2.plot(df['d_t'],df['ST4'], color='blue',label='Soil Temperature 4 -27.5 cm [°C]')
ax2.plot(df['d_t'],df['ST5'], color='magenta',label='Soil Temperature 5 -7.5 cm [°C]')
ax2.plot(df['d_t'],df['ST6'], color='grey',label='Soil Temperature 6 -27.5 cm [°C]')
fig2.legend(bbox_to_anchor=(0.8,0.45,0.5,0.5))
```
