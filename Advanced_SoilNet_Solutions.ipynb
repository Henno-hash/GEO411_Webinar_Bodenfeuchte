### Definieren einer Funktion  
  
Um größere Datensätze mit mehreren Datein automatisiert vorzubereiten und die gewünschten Aufnahmezeiträume zu extrahieren, sollte eine Funktion aus den bisherig gelernten Arbeitsschritten erstellt werden, welche den entsprechenden Datensatz der Orbit-Richtung übergibt.  
  
Es gilt die Variable *day* hervorzuheben. Hierbei handelt es sich um den spezifischen Tag. Dieser wird im Namen der Datei angegeben und muss über die dementsprechenden Stellennummern im Verzeichnisstring extrahiert werden.
  
*day* = file[-8:-4] + '-' + file[-14:-12] + '-' + file[-11:-9]
  
Die Variable ist notwendig, da die Filter den spezifischen Tag benötigen. Der Tag ändert sich allerdings mit jeder geöffneten Datei.

```python
import glob
import pandas as pd
import datetime
import io
```

Die Funktion *data_aggregation_time_series(inDir, orb_dir, day)* dient zur automatisierten Berechnung der durchschnittlichen Bodenfeuchte und Bodentemperatur der jeweiligen Messsonden für den Aufnahmezeitpunkt einer Sentinel-1 Szene.
  
Diese Funktion kann dann auf den Zieldatensatz angewendet werden. Zu definieren sind der eigentliche Pfad der Datensätze, die Orbit-Richtung und *day*.  
  
Je nach Orbit-Richtung wird der entsprechende Data-Frame zurückgegeben.

```python
def data_aggregation_time_series(inDir, orb_dir, day):
    file = inDir
    df=pd.read_csv(file, sep = '\t')

    df = df.append(pd.Series(), ignore_index=True)
    df.iloc[-1] = df.columns
    df.columns=['date','time','node','bat','V5','V6','V7','V8','EPS1','SM1','ST1','V12','V13','EPS2',
                'SM2','ST2','V17','V18','EPS3','SM3','ST3','V22','V23','EPS4','SM4','ST4','V27','V28',
                'EPS5','SM5','ST5','V32','V33','EPS6','SM6','ST6','V37','WPOT7','MPS_2_1_T7','V40','V41',
                'V42','WPOT8','MPS_2_2_T8','V45','V46','V47','WPOT9','MPS_2_3_T9','V50','V51','d_t']
    
    df.SM1 = df.SM1.astype(float)
    df.SM2 = df.SM2.astype(float)
    df.SM3 = df.SM3.astype(float)
    df.SM4 = df.SM4.astype(float)
    df.SM5 = df.SM5.astype(float)
    df.SM6 = df.SM6.astype(float)
    df.ST1 = df.ST1.astype(float)
    df.ST2 = df.ST2.astype(float)
    df.ST3 = df.ST3.astype(float)
    df.ST4 = df.ST4.astype(float)
    df.ST5 = df.ST5.astype(float)
    df.ST6 = df.ST6.astype(float)

    df.drop(df.columns[37:51], inplace = True, axis = 1)

    df.sort_values(df.columns[1], inplace = True)

    df_2 = df[['d_t', 'SM1', 'ST1', 'SM2', 'ST2', 'SM3', 'ST3', 'SM4', 'ST4', 'SM5', 'ST5', 'SM6', 'ST6']]

    df_2 = df_2.assign(New_Column = ((df_2['SM1'] + df_2['SM3'] + df_2['SM5'])/3))
    df_2.rename(columns={'New_Column':'SM -7.5 cm'}, inplace = True)
    df_2 = df_2.assign(New_Column = ((df_2['SM2'] + df_2['SM4'] + df_2['SM6'])/3))
    df_2.rename(columns={'New_Column':'SM -27.5 cm'}, inplace = True)
    df_2 = df_2.assign(New_Column = ((df_2['ST1'] + df_2['ST3'] + df_2['ST5'])/3))
    df_2.rename(columns={'New_Column':'ST -7.5 cm'}, inplace = True)
    df_2 = df_2.assign(New_Column = ((df_2['ST2'] + df_2['ST4'] + df_2['ST6'])/3))
    df_2.rename(columns={'New_Column':'ST -27.5 cm'}, inplace = True)

    df_2.drop(df_2.columns[1:13], inplace = True, axis = 1)

    if orb_dir == 'asc':

        filt_asc=((df_2['d_t']> (day +' 16:35:00'))&(df_2['d_t']< (day + ' 17:15:00')))
        df_asc = df_2.loc[filt_asc]

        rows, columns = df_asc.shape

        df_asc = df_asc.append(pd.Series(dtype='float64'), ignore_index=True)    
        df_asc.loc[rows] = df_asc.mean()
        df_asc.loc[rows, ['d_t']] = (day +' 17:00:00')
        df_asc['d_t'] = pd.to_datetime(df_asc['d_t'])
        df_asc = df_asc.drop(df_asc.index[0:(rows)])
        
        return df_asc
        
    elif orb_dir == 'desc':
        
        filt_desc = ((df_2['d_t']> (day +' 04:59:59'))&(df_2['d_t']< (day +' 05:36:01')))
        df_desc = df_2.loc[filt_desc]

        rows, columns = df_desc.shape

        df_desc = df_desc.append(pd.Series(dtype='float64'), ignore_index=True)    
        df_desc.loc[rows] = df_desc.mean()
        df_desc.loc[rows, ['d_t']] = (day + ' 05:25:00')
        df_desc['d_t'] = pd.to_datetime(df_desc['d_t'])
        df_desc = df_desc.drop(df_desc.index[0:(rows)])

        return df_desc
```

In der Funktion *auto_asc_df(inDir)* wird mittels Verschachtelung die Funktion *data_aggregation_time_series(file, orb_dir, day)* auf alle Textdatein innerhalb des Zielordners angewandt. Nach dem ersten Durchlauf wird ein Data-Frame erstellt (*all_df*), welcher nur die aggregierten Daten enthält. Mit jeden weiteren Durchlauf wird dieser gefüllt.
  
Zunächst für die aufsteigende (ascending) Orbit-Richtung: *asc_all_df*

```python
import glob 
inDir = 'E:/UNI/GEO411_Bodenfeuchte/1_Github/Datensatz_Adv_NB/'
def auto_asc_df(inDir):
    files = sorted(glob.glob(inDir + '*.txt', recursive=False))
    orb_dir = 'asc'
    i=0
    for file in files:
        day = file[-8:-4] + '-' + file[-14:-12] + '-' + file[-11:-9]
        #print(file) #Debug falls in einem File ein Fehler vorliegt
        i+=1
        df = data_aggregation_time_series(file, orb_dir, day) 
        if (i == 1):      
            all_df = df
        else:        
            all_df = pd.concat([all_df, df])
    asc_all_df = all_df
    return asc_all_df
asc_all_df=auto_asc_df(inDir)
```

Und dann für die absteigende (descending) Orbit-Richtung: *desc_all_df*

```python
import glob        
inDir = 'E:/UNI/GEO411_Bodenfeuchte/1_Github/Datensatz_Adv_NB/'
def auto_desc_df(inDir):
    files = sorted(glob.glob(inDir + '*.txt', recursive=False))
    orb_dir = 'desc'
    i=0
    for file in files:
        day = file[-8:-4] + '-' + file[-14:-12] + '-' + file[-11:-9]
        #print(file) #Debug falls in einem File ein Fehler vorliegt
        i+=1
        df = data_aggregation_time_series(file, orb_dir, day) 
        if (i == 1):      
            all_df = df
        else:        
            all_df = pd.concat([all_df, df])
    desc_all_df = all_df
    return desc_all_df
desc_all_df=auto_desc_df(inDir)
```

### Plotten der Ergebnisse

Plotten der durchschnittlichen Bodentemperatur und Bodenfeuchte für jeweils auf- und absteigende Orbit-Richtung.
  
Das Plotten kann zwar mit in die eigentlichen Funktionen *auto_asc_df* und *auto_desc_df* integriert werden, hiervon würden wir allerdings abraten, da je nach Datengröße (Zeitreihe) die Plots manuell angepasst werden müssen. Automatisierte Lösungsansätze mit Matplotlib oder anderen Packages wäre jedoch auch denkbar. 

Die Funktion *plot_time_series(orb_dir, asc_all_df, desc_all_df)* plottete die aggregierte Bodenfeuchte und Bodentemperatur der übergegeben Data-Frames *asc_all_df* und *desc_all_df* je nach Orbit-Richtung und übergibt die Diagramme.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MultipleLocator

def plot_time_series(orb_dir, asc_all_df, desc_all_df):
    fig1, ax1 = plt.subplots(figsize=(10,5))

    if orb_dir == 'asc':
        ax1.set_title('Soil Moisture and Soil Temperature (01.05.2018-13.05.2018) for Ascending Orbit', fontweight='bold')
        all_df = asc_all_df
    elif orb_dir == 'desc':
        ax1.set_title('Soil Moisture and Soil Temperature (01.05.2018-13.05.2018) for Descending Orbit', fontweight='bold')
        all_df = desc_all_df
    
    ax1.set_xlabel('Date [YYYY-MM-DD HH:MM:SS]')
    ax1.set_ylabel('Soil Moisture [%]')
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    ax1.xaxis.set_major_locator(MultipleLocator(2.5))
    ax1.plot(asc_all_df['d_t'],all_df['SM -7.5 cm'],color='blue', label='Mean Soil Moisture -7.5 cm [%]')
    ax1.plot(asc_all_df['d_t'],all_df['SM -27.5 cm'],color='darkblue', label='Mean Soil Moisture -27.5 cm [%]')


    ax2=ax1.twinx()
    ax2.set_ylabel('Soil Temperature [°C]')
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    ax2.xaxis.set_major_locator(MultipleLocator(2.5))
    ax2.plot(asc_all_df['d_t'],all_df['ST -7.5 cm'], color='red',label='Mean Soil Temperature -7.5 cm [°C]')
    ax2.plot(asc_all_df['d_t'],all_df['ST -27.5 cm'], color='darkred',label='Mean Soil Temperature -27. cm [°C]')

    fig1.legend(bbox_to_anchor=(0.8,0.45,0.5,0.5))
    
    return fig1
```

```python
fig3 = plot_time_series('asc', asc_all_df, desc_all_df)
fig4 = plot_time_series('desc', asc_all_df, desc_all_df)
```
